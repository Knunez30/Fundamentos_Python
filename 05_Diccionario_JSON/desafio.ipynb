```python
# Tu código aquí — exploración libre
import json
import os
ARCHIVO = "clinica_s08_avanzado.json"
if os.path.exists(ARCHIVO):
    with open(ARCHIVO, "r", encoding="utf-8") as f:
        datos_clinica = json.load(f)
else:    datos_clinica = []
print(f"Dataset listo: {len(datos_clinica)} pacientes")
print(f"Campos disponibles: {list(datos_clinica[0].keys())}")
print()
print("--- Primer registro ---")
for k, v in datos_clinica[0].items():
    print(f"  {k}: {v}")
```

```python
# Sigue explorando...
# De esos 57 pacientes, ¿cuántos son hombres y cuántos mujeres? ¿Cuál es la edad promedio? 
hombres = 0
mujeres = 0
suma_edades = 0

for paciente in datos_clinica:
    suma_edades += paciente["edad"]
    if paciente["genero"] == "M":
        hombres += 1
    else:
        mujeres += 1
print("*" * 20)
print(" Cantidad de Pacientes por Género ")
print(f"Hombres: {hombres}")
print(f"Mujeres: {mujeres}")
print(f"Edad promedio: {suma_edades / len(datos_clinica):.2f}")

# ¿Cuáles son las enfermedades más comunes? 
enfermedades = {}
for paciente in datos_clinica:
    enfermedad = paciente["enfermedad"]
    enfermedades[enfermedad] = enfermedades.get(enfermedad, 0) + 1
print("*" * 20)
print("Enfermedades más comunes:")
for enfermedad, count in sorted(enfermedades.items(), key=lambda x: x[1], reverse=True):
    print(f"  {enfermedad}: {count}")

# ¿Qué medicamento se receta con más frecuencia? 
medicamentos = {}
for paciente in datos_clinica:
    medicamento = paciente["medicamento"]
    medicamentos[medicamento] = medicamentos.get(medicamento, 0) + 1
print("*" * 20)
print("Medicamento más recetado:")
for medicamento, count in sorted(medicamentos.items(), key=lambda x: x[1], reverse=True):
    print(f"  {medicamento}: {count}")
```

```python
# Más análisis...
# 1. ¿Cuántas visitas han tenido en promedio?

total_visitas = sum(p["visitas"] for p in datos_clinica)
promedio_visitas = total_visitas / len(datos_clinica)

print("*" * 20)
print(f"Visitas promedio por paciente: {promedio_visitas:.2f}")

# 2. Paciente con MÁS visitas
paciente_mas_visitas = max(datos_clinica, key=lambda x: x["visitas"])
print("*" * 20)
print(f"Paciente con más visitas: {paciente_mas_visitas['nombre']} ({paciente_mas_visitas['visitas']} visitas)")

# 3. Paciente con MENOS visitas
paciente_menos_visitas = min(datos_clinica, key=lambda x: x["visitas"])
print("*" * 20)
print(f"Paciente con menos visitas: {paciente_menos_visitas['nombre']} ({paciente_menos_visitas['visitas']} visitas)")

# 4. Paciente más JOVEN
paciente_mas_joven = min(datos_clinica, key=lambda x: x["edad"])
print("*" * 20)
print(f"Paciente más joven: {paciente_mas_joven['nombre']} ({paciente_mas_joven['edad']} años)")

# 5. Paciente más VIEJO
paciente_mas_edad = max(datos_clinica, key=lambda x: x["edad"])
print("*" * 20)
print(f"Paciente de mayor edad: {paciente_mas_edad['nombre']} ({paciente_mas_edad['edad']} años)")
```

```python

# ¿Cuántos pacientes tienen activo = true? ¿Cuántos pacientes tienen activo = false?
activos = sum(1 for paciente in datos_clinica if paciente["activo"])
inactivos = len(datos_clinica) - activos
print("*" * 20)
print(f"Pacientes activos: {activos}")
print(f"Pacientes inactivos: {inactivos}")

#Cantón de cada paciente
cantones = {}
for paciente in datos_clinica:
    canton = paciente["canton"]
    cantones[canton] = cantones.get(canton, 0) + 1
print("*" * 20)
print("Cantidad de pacientes por cantón:")
for canton, count in sorted(cantones.items(), key=lambda x: x[1], reverse=True):
    print(f"  {canton}: {count}")
```

```python
#más análisis...
def analizar_frecuencia(lista_datos, llave):
    conteo = {}
    for paciente in lista_datos:
        valor = paciente[llave]
        conteo[valor] = conteo.get(valor, 0) + 1
    return conteo

# 2. lista con las categorías que se quieren analizar
categorias = ["enfermedad", "medicamento", "canton"]

# 3. Bucle Anidado
# Un for que recorre las categorías y el otro (dentro de la función) recorre los pacientes
print("--- ANÁLISIS DE FRECUENCIA ---")
for cat in categorias:
    resultado = analizar_frecuencia(datos_clinica, cat)
    
    # Ordenamos y mostramos el top 3 de cada una
    top_3 = sorted(resultado.items(), key=lambda x: x[1], reverse=True)[:3]
    
    print(f"\nTop 3 en {cat.upper()}:")
    for nombre, cantidad in top_3:
        print(f" - {nombre}: {cantidad}")
```
