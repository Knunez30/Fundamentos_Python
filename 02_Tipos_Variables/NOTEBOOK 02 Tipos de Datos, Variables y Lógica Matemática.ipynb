## 1. Tipos de Datos y Variables en el Mundo Real 📦

En lugar de usar variables abstractas como `x` o `y`, los profesionales de Python nombramos nuestras variables según el contexto del negocio (PEP 8 exige nombres descriptivos en minúsculas separados por guiones bajos: `snake_case`).

| Tipo de Dato | Propósito | Ejemplos en un E-commerce |

| **`bool`** (Booleano) | Interruptores lógicos (Sí/No) | `True`, `False` |

```python
# 💡 Live Coding: Inicializando el Inventario de una Tienda

# 1. Crea las variables correspondientes con sus datos y aplica PEP 8
nombre_producto = "monitor Dell 31\""
precio_unitario =  350000.50
unidades_en_stock = 15
aplica_envio_gratis = True

# 2. Usa print() y type() para auditar cómo Python clasificó cada variable
print("--- AUDITORÍA DE TIPOS DE DATOS ---")

# Escribe tu código aquí:
print(f'Producto: {type(nombre_producto)}')
print(f'Precio: {type(precio_unitario)}')
print(f'Unidades en stock:{type(unidades_en_stock)}')
print(f' Envío Gratis {type (aplica_envio_gratis)}')
```

## 2. Entrada de Datos y Casting Seguro 🛡️

La función `input()` es nuestra ventana para interactuar con el usuario. 
⚠️ **¡Regla de Oro!** Todo lo que entra por `input()` es un texto (`str`). Si el usuario escribe `5`, Python lo ve como la letra `'5'`, no como el número `5`.

Para hacer matemáticas, debemos "traducir" (Casting) usando `int()` o `float()`. ¡Pero cuidado! Si intentamos convertir la palabra "diez" a número, el programa explotará. Por eso recordaremos el `try-except` de nuestra primera clase.

```python
# 💡 Live Coding: Captura y Conversión de Cantidad (Con Seguridad)

print("Bienvenido al Sistema de Ventas")

# 1. Solicita al usuario cuántas unidades desea comprar
cantidad_str = input(f"Cuántas unidades de '{nombre_producto}' desea comprar?")

# 2. Escribe un bloque try-except para convertir cantidad_str a entero (int)
# Asegúrate de validar si la cantidad es mayor a 0.

# Escribe tu código aquí:

try:
    cantidad = (int(cantidad_str))
    print(type(cantidad))
    if cantidad <1:
        print("❌Error: Ingrese una cantidad válida, mayor a cero")
    else:
        print(f"Procesando {cantidad} unidades ....")
        print(type(cantidad))
except ValueError:
    print("❌Error crítico: Por favor ingrese un número entero válido.")
    cantidad = 0
```

## 3. Operadores Matemáticos: El Calculador de Pedidos 🧮

Ya tenemos números reales en la memoria. Ahora podemos usar la calculadora interna de Python:

* `+` (Suma), `-` (Resta), `*` (Multiplicación), `/` (División Exacta)
* `//` **(División Entera):** Descarta los decimales. Ideal para saber cuántas cajas llenas podemos armar.
* `%` **(Módulo):** Nos da el *sobrante* de una división. Clave para saber si un número es par (`numero % 2 == 0`).

```python
# 💡 Live Coding: Cálculo Completo del Tícket y Logística

TASA_IMPUESTO = 0.13  # 13% IVA (Constantes en mayúscula según PEP 8)
CAPACIDAD_CAJA = 4    # Cada caja de envío soporta 4 monitores


# 1. Matemáticas Financieras
# Calcula el subtotal, el monto_impuesto y el total_final
subtotal = precio_unitario * cantidad
monto_impuesto = subtotal * TASA_IMPUESTO
total_final = subtotal + monto_impuesto

# 2. Matemáticas Logísticas 
# Usa // para saber cuántas cajas completas salen
# Usa % para saber cuántos monitores sobran sueltos
cajas_llenas = cantidad // CAPACIDAD_CAJA
monitores_sueltos = cantidad % CAPACIDAD_CAJA

# 3. Impresión Profesional con F-Strings
print("\n" + "="*30)
print("🧾 TÍCKET DE COMPRA")
print("="*30)
# Imprime los resultados con 2 decimales y formato de miles (,)
print(f"Subtotal: ₡ {subtotal:,.2f}")
print(f"IVA {TASA_IMPUESTO:.0%}: ₡{monto_impuesto:,.2f}")
print(f"Total: ₡ {total_final:,.2f}")
print("-" * 30)
print("📦 Logística de Envío")
print(f"Cajas completas requeridas: {cajas_llenas}")
print(f"Empaque individual: {monitores_sueltos}")
print("-" * 30)
```

## 4. Operadores de Comparación y Lógicos: Lógica de Negocio 🧠

Los **operadores de comparación** (`>`, `<`, `==`, `!=`, etc.) generan respuestas de `True/False`.
Los **operadores lógicos** (`and`, `or`, `not`) combinan esas respuestas para crear el "cerebro" de nuestra aplicación.

| Operador Lógico | Regla Mnemotécnica |
|:----------------|:-------------------|
| `and` (Y)       | **El Exigente:** TODAS las condiciones deben ser `True`. |
| `or` (O)        | **El Flexible:** Con que UNA condición sea `True`, basta. |
| `not` (NO)      | **El Rebelde:** Invierte el valor (`not True` es `False`). |

```python
# 💡 Live Coding: Motor de Promociones y Descuentos

es_cliente_vip = False

# PROMOCIÓN 1: Descuento del 10% (Uso de AND)
# Regla: Aplica SI el total es mayor a 1.000.000 Y lleva más de 20 unidades.
aplica_descuento_10 = (total_final > 1000000) and (cantidad > 20)

# PROMOCIÓN 2: Regalo Sorpresa (Uso de OR)
# Regla: Aplica SI es Cliente VIP O SI compró exactamente 5 unidades.
aplica_regalo = es_cliente_vip or (cantidad == 5) 

print("\n🎯 ANÁLISIS DE PROMOCIONES:")
# Imprime las dos variables de promociones aquí:
print(f"Se aplica el descuento 10%: {aplica_descuento_10}")
print(f"Lleva regalo: {aplica_regalo}")
```
